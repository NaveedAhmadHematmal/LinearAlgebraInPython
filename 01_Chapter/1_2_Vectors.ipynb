Lines drawing

```python
from matplotlib import pyplot as pl

pl.grid()
pl.plot([1, 2], [3, 3])
pl.plot([1, 1.7], [2.5, 2.5])
pl.plot([1, 1], [1.5, 3])
```

```python
from matplotlib import pyplot as pl

pl.grid()
pl.plot([1, 1.5], [1, 3])
pl.plot([1.5, 2], [3, 1])
pl.plot([1.19, 1.82], [1.75, 1.75])
```

Vectors Drawing (Standard Position)

```python
from matplotlib import pyplot as plt
import numpy as np

v1 = np.array([1, 2])
plt.plot([0, v1[0]], [0, v1[1]])
plt.grid()
```

Equal vectors

```python
from matplotlib import pyplot as plt
import numpy as np

v1 = np.array([-1, -1])
v2 = np.array([1, -2])
v3 = np.array([2, 0])

plt.plot([-2, v1[0]], [1, v1[1]])
plt.plot([0, v2[0]], [0, v2[1]])
plt.plot([1, v3[0]], [2, v3[1]])

plt.grid()
```

```python
from matplotlib import pyplot as plt
import numpy as np

v1 = np.array([1, 3])
v2 = np.array([3, 1])

plt.plot([0, v1[0]], [0, v1[1]])
plt.plot([0, v2[0]], [0, v2[1]])

plt.grid()
```

Column vs Row vectors

```python
from matplotlib import pyplot as plt
import numpy as np

columnV = np.array([[1], [3]])
rowV = np.array([1, 3])

plt.grid()

plt.plot([0, columnV[0]], [0, columnV[1]])
plt.plot([0, rowV[0]], [0, rowV[1]])

```

3d vector

```python
from matplotlib import pyplot as plt
import numpy as np

v1 = np.array([1, 3, 5])

plt.plot([0, v1[0]], [0, v1[1]], [0, v1[2]])

plt.grid()
```

Transpose a Vector

```python
import numpy as np

v = np.array([[1, 2, 3]])
v1 = v.T
print("Row vector")
print(v)
print("transposed")
print(v1)
```

Vectors addition

```python
import numpy as np

v1 = np.array([[1, 2, 3]])
v2 = np.array([[3, 2, 3]])
v = v1 + v2
print(v)
```

Vector subtraction

```python
import numpy as np

v1 = np.array([[1, 2, 3]])
v2 = np.array([[3, 2, 3]])
v = v1 - v2
print(v)
```

Vector Scaling or multiplication with a scalar

```python
import numpy as np

v1 = np.array([[1, 2, 3]])
v = (-2) * v1
print(v)
```
